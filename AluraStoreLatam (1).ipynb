#1. Análisis de facturación



### Calcular facturación total (precio * cantidad, aunque no veo columna cantidad, asumiremos que cada fila es una venta)


```python
print("-----------------------------TIENDA 01-----------------------------")
tienda['Facturación'] = tienda['Precio'] + tienda['Costo de envío']
display(tienda['Facturación'])
print("-----------------------------TIENDA 02-----------------------------")
tienda2['Facturación'] = tienda2['Precio'] + tienda2['Costo de envío']
display(tienda2['Facturación'])
print("-----------------------------TIENDA 03-----------------------------")
tienda3['Facturación'] = tienda3['Precio'] + tienda3['Costo de envío']
display(tienda3['Facturación'])
print("-----------------------------TIENDA 04-----------------------------")
tienda4['Facturación'] = tienda4['Precio'] + tienda4['Costo de envío']
display(tienda4['Facturación'])
```

### Facturación total


```python
print("-----------------------------TIENDA 01-----------------------------")
facturacion_total = tienda['Facturación'].sum()
print(f"\nFacturación total: ${facturacion_total:,.2f}")
print("-----------------------------TIENDA 02-----------------------------")
facturacion_total = tienda2['Facturación'].sum()
print(f"\nFacturación total: ${facturacion_total:,.2f}")
print("-----------------------------TIENDA 03-----------------------------")
facturacion_total = tienda3['Facturación'].sum()
print(f"\nFacturación total: ${facturacion_total:,.2f}")
print("-----------------------------TIENDA 04-----------------------------")
facturacion_total = tienda4['Facturación'].sum()
print(f"\nFacturación total: ${facturacion_total:,.2f}")
```

## Facturación promedio por venta

```python
print("-----------------------------TIENDA 01-----------------------------")
facturacion_total = tienda['Facturación'].mean()
print(f"\nFacturación promedio: ${facturacion_total:,.2f}")
print("-----------------------------TIENDA 02-----------------------------")
facturacion_total = tienda2['Facturación'].mean()
print(f"\nFacturación promedio: ${facturacion_total:,.2f}")
print("-----------------------------TIENDA 03-----------------------------")
facturacion_total = tienda3['Facturación'].mean()
print(f"\nFacturación promedio: ${facturacion_total:,.2f}")
print("-----------------------------TIENDA 04-----------------------------")
facturacion_total = tienda4['Facturación'].mean()
print(f"\nFacturación promedio: ${facturacion_total:,.2f}")
```

## Facturación por tienda

```python
tienda_compl = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
facturacion_por_todas_tienda = tienda_compl.groupby('Producto')['Facturación'].sum().sort_values(ascending=False)

print("\nFacturación de todas las tiendas:")
display(facturacion_por_todas_tienda)
print("----------------------------------------")
facturacion_por_tienda = tienda.groupby('Producto')['Facturación'].sum().sort_values(ascending=False)
print("\nFacturación por tienda1:")
display(facturacion_por_tienda)
print("----------------------------------------")
facturacion_por_tienda2 = tienda2.groupby('Producto')['Facturación'].sum().sort_values(ascending=False)
print("\nFacturación por tienda2:")
display(facturacion_por_tienda2)
print("----------------------------------------")
facturacion_por_tienda3 = tienda3.groupby('Producto')['Facturación'].sum().sort_values(ascending=False)
print("\nFacturación por tienda3:")
display(facturacion_por_tienda3)
print("----------------------------------------")
facturacion_por_tienda4 = tienda4.groupby('Producto')['Facturación'].sum().sort_values(ascending=False)
print("\nFacturación por tienda4:")
display(facturacion_por_tienda4)
print("----------------------------------------")
```

# 6. Envío promedio por tienda

```python
envio_promedio_todas_tienda = round(tienda_compl['Costo de envío'].mean(),1)
print("\nFacturación de todas las tiendas:")
display(envio_promedio_todas_tienda)
print("----------------------------------------")
envio_promedio_por_tienda = round(tienda['Costo de envío'].mean(),1)
print("\nEnvío promedio por tienda1:")
display(envio_promedio_por_tienda)
print("----------------------------------------")
envio_promedio_por_tienda2 = round(tienda2['Costo de envío'].mean(),1)
print("\nEnvío promedio por tienda2:")
display(envio_promedio_por_tienda2)
print("----------------------------------------")
envio_promedio_por_tienda3 = round(tienda3['Costo de envío'].mean(),1)
print("\nEnvío promedio por tienda3:")
display(envio_promedio_por_tienda3)
print("----------------------------------------")
envio_promedio_por_tienda4 = round(tienda4['Costo de envío'].mean(),1)
print("\nEnvío promedio por tienda4:")
display(envio_promedio_por_tienda4)
print("----------------------------------------")
envio_promedio_todas_tienda_l = [envio_promedio_por_tienda,envio_promedio_por_tienda2,envio_promedio_por_tienda3,envio_promedio_por_tienda4]
envio_promedio_todas_tienda_l
plt.figure(figsize=(10, 6))
plt.title("Envío promedio por tienda")
plt.bar(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'], envio_promedio_todas_tienda_l, color=sns.palettes.mpl_palette('Dark2'))

# Convert the list to a pandas Series for plotting
envio_promedio_series = pd.Series(envio_promedio_todas_tienda_l, index=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])

plt.figure(figsize=(8, 8))
envio_promedio_series.plot(kind='pie', autopct='%1.1f%%', startangle=90, labels=envio_promedio_series.index)
plt.title('Distribución % de Envío Promedio por Tienda', fontsize=14)
plt.ylabel('')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(x=envio_promedio_series.index, y=envio_promedio_series.values,
            s=envio_promedio_series.values/100, alpha=0.7) # Adjusted marker size for better visualization
plt.title('Costo de envio por tienda', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
